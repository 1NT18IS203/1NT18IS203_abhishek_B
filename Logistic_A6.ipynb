```python
import pandas as pd
import math
import numpy as np
from sklearn.model_selection import train_test_split

#reading dataset
dataset=pd.read_csv("Student_University.csv")
print(dataset)

#splitting dataset into train and test
X=dataset.values[:,0:2]
Y=dataset.values[:,2]
# print(X)
print(Y)

X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.20)

#intializing coefficients
b0=0.0
b1=0.0
b2=0.0
x1=X_train[:,0]
x2=X_train[:,1]
# print(x1)


i=0
alpha=0.001
while(i<10000):
    for j in range(len(x1)):
        p=(1/(1+np.exp(-(b0+b1*x1[j]+b2*x2[j]))))
        b0=b0 + alpha * (y_train[j] - p) * p* (1 - p) * 1.0
        b1=b1+alpha*(y_train[j]-p)*p*(1-p)*x1[j]
        b2=b2+alpha*(y_train[j]-p)*p*(1-p)*x2[j]
    i+=1


print("b0",b0)
print("b1",b1)
print("b2",b2)


xtest1=X_test[:,0]
xtest2=X_test[:,1]
y_pred=[]
for k in range(len(xtest1)):
    y_pred.append(np.round(1/(1+np.exp(-(b0+b1*xtest1[k]+b2*xtest2[k])))))
print(y_pred)
print(y_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```
